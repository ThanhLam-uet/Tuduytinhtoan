```python
def diemnamtrong(quad,p):
    x,y=p 
    n=len(quad)
    inside=False
    for i in range(n):
        x1,y1=quad[i]
        x2,y2=quad[(i+1)%4]
        if min(y1,y2)<=y<max(y1,y2) and min(x1,x2)<=x<=max(x1,x2):
            if (y2-y1)*(x-x1)==(y-y1)*(x2-x1):
                return True
        if min(y1,y2)<=y<=max(y1,y2):
            if x<max(x1,x2):
                if x1!=x2:
                    x_giaodiem=(y2-y1)*(x2-x1)/(y-y1)+x1
                else:
                    x_giaodiem=x1 
                    inside=not inside
    return inside
            
            

```
