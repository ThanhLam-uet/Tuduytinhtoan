```python
# Bài 2: Tính diện tích trồng cây (Khu đất a x b, hình tròn ở giữa) 
def bai_2():
    a = float(input("Nhập chiều rộng a: "))
    b = float(input("Nhập chiều dài b: "))
    pi = 3.14 
    r = a / 2
    s_hcn = a * b
    s_tron = pi * (r ** 2)
    s_con_lai = s_hcn - s_tron
    print(f"Diện tích trồng cây là: {s_con_lai:.2f}") 
```

```python
# Bài 5: In chữ cái thường liền trước của chữ cái hoa A tương ứng [cite: 11]
def bai_5():
    c = input("Nhập chữ cái hoa: ")
    if c == 'A':
        print("Trường hợp đặc biệt")
    else:
        char_thuong = c.lower()
        res = chr(ord(char_thuong) - 1)
        print(res)
```

```python
# Bài 6: Kiểm tra tam giác và tính diện tích [cite: 12, 13]
import math 
def bai_6():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    c = float(input("Nhập c: "))
    if a + b > c and a + c > b and b + c > a:
        p = (a + b + c) / 2
        s = math.sqrt(p * (p - a) * (p - b) * (p - c))
        print(f"Diện tích tam giác: {s:.1f}") 
    else:
        print("Khong phai 3 canh cua tam giac") 
```

```python
# Bài 7: In kí tự thứ 5 và thứ 9 của chuỗi >= 20 kí tự [cite: 14]
def bai_7():
    s = input("Nhập chuỗi (độ dài >= 20): ")
    if len(s) >= 20:
        print(f"Kí tự thứ 5: {s[4]}, thứ 9: {s[8]}") 
    else:
        print("Chuỗi quá ngắn")
```

```python
# Bài 8: Tính tiền điện [cite: 15, 16, 17, 18, 19, 20, 21, 22]
def bai_8():
    ten = input("Nhập họ tên chủ hộ: ")
    truoc = int(input("Chỉ số điện kế tháng trước: "))
    nay = int(input("Chỉ số điện kế tháng này: "))
    kwh = nay - truoc
    tien = 0
    
    if kwh <= 50:
        tien = kwh * 1984 
    elif kwh <= 100:
        tien = 50 * 1984 + (kwh - 50) * 2050 
    elif kwh <= 200:
        tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380 
    elif kwh <= 300:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998 
    elif kwh <= 400:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350 
    else:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460
    
    tong_tien = tien * 1.08 
    print(f"Ho va ten: {ten}") [cite: 23]
    print(f"Tien phai tra la: {round(tong_tien)}") 
```

```python
# 9. Chuyển đổi số thành chữ (dùng match case) 
def extra_9():
    n = int(input("Nhập số (0-9): "))
    match n:
        case 0: print("Không")
        case 1: print("Một")
        case 2: print("Hai")
        case 3: print("Ba")
        case 4: print("Bốn")
        case 5: print("Năm")
        case 6: print("Sáu")
        case 7: print("Bảy")
        case 8: print("Tám")
        case 9: print("Chín")
        case _: print("Ngoài phạm vi")
```
