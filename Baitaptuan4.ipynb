```python
# Bài 1: In các số dương lẻ bé hơn 100
def bai_1():
    for i in range(1, 100, 2): 
        print(i, end=" ")

# Bài 2: In các số dương chẵn chia hết cho 3 bé hơn 100
def bai_2():
    for i in range(1, 100): 
        if i % 2 == 0 and i % 3 == 0:
            print(i, end=" ")

# Bài 3: Bảng cửu chương của a
def bai_3():
    a = int(input("Nhập a: ")) 
    for i in range(1, 11):
        print(f"{a} x {i} = {a * i}")

# Bài 4: Vẽ tam giác cân độ cao h
def bai_4():
    h = int(input("Nhập độ cao h: ")) 
    for i in range(1, h + 1):
        for j in range(h - i):
            print(" ", end="")
        for k in range(2 * i - 1):
            print("*", end="")
        print() 
```

```python
import math

# Bài 5: Tính tổng S = 1 + 2 + ... + n
def bai_5():
    n = int(input("Nhập n: ")) 
    s = 0
    for i in range(1, n + 1):
        s += i
    print(f"S = {s}")

# Bài 6 & 7: Liệt kê và đếm ước dương của a
def bai_6_7():
    a = int(input("Nhập a: "))
    count = 0
    print("Các ước của a:", end=" ")
    for i in range(1, a + 1):
        if a % i == 0:
            print(i, end=" ")
            count += 1
    print(f"\nSố lượng ước: {count}")

# Bài 8: Ước chung của a và b
def bai_8():
    a = int(input("Nhập a: ")) 
    b = int(input("Nhập b: "))
    limit = a if a < b else b
    print(f"Ước chung của {a} và {b}:", end=" ")
    for i in range(1, limit + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")

# Bài 9: Kiểm tra số chính phương
def bai_9():
    a = int(input("Nhập a: "))
    if a < 0:
        print("Không phải số chính phương")
        return
    can = int(math.sqrt(a))
    if can * can == a:
        print(f"{a} là số chính phương")
    else:
        print(f"{a} không phải số chính phương")

# Bài 10: Kiểm tra số hoàn hảo
def bai_10():
    a = int(input("Nhập a: "))
    tong_uoc = 0
    for i in range(1, a):
        if a % i == 0:
            tong_uoc += i
    if tong_uoc == a:
        print(f"{a} là số hoàn hảo")
    else:
        print(f"{a} không phải là số hoàn hảo")
```

```python
# Bài 11: Nhập cho đến khi là số dương
def bai_11():
    while True:
        a = int(input("Nhập số nguyên a: ")) 
        if a > 0:
            print("Bạn nhập đúng quy tắc") 
            break

# Bài 12: Tìm k sao cho S(k) = 1+2+...+k <= n lớn nhất
def bai_12():
    n = int(input("Nhập n: "))
    s = 0
    k = 0
    while s + (k + 1) < n:
        k += 1
        s += k
    print(f"k lớn nhất là {k} với S({k}) = {s}")

# Bài 13: Tìm n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n > A
def bai_13():
    a = float(input("Nhập A: ")) 
    s = 0
    n = 0
    while s <= a:
        n += 1
        s += 1/n
    print(f"n nhỏ nhất là {n}")

# Bài 14: Tìm Max/Min trong dãy số dừng bởi -1 (Không dùng mảng)
def bai_14():
    print("Nhập dãy số (nhập -1 để dừng):") 
    n = int(input())
    if n == -1: return
    
    max_val = min_val = n
    while True:
        n = int(input())
        if n == -1: break 
        if n > max_val: max_val = n 
        if n < min_val: min_val = n 
    
    print(f"Số lớn nhất: {max_val}")
    print(f"Số nhỏ nhất: {min_val}")
```

```python
# Bài 15, 16, 17: Đếm và tính tổng chữ số
def bai_digits():
    n = int(input("Nhập số nguyên dương n: "))
    count = 0
    count_even = 0
    count_odd = 0
    total_sum = 0
    temp = n
    
    while temp > 0:
        digit = temp % 10
        total_sum += digit 
        count += 1 
        if digit % 2 == 0:
            count_even += 1 
        else:
            count_odd += 1 
        temp //= 10
        
    print(f"Số chữ số: {count}")
    print(f"Chữ số chẵn: {count_even}, Chữ số lẻ: {count_odd}")
    print(f"Tổng các chữ số: {total_sum}")

# Bài 18: Kiểm tra dạng 3^k
def bai_18():
    n = int(input("Nhập n: "))
    if n <= 0:
        print("Không phải dạng 3^k")
        return
    temp = n
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        print(f"{n} có dạng 3^k")
    else:
        print(f"{n} không phải dạng 3^k")

# Bài 19: Fibonacci lớn nhất không vượt quá A
def bai_19():
    a = int(input("Nhập A: ")) 
    f1, f2 = 1, 1
    if a < 1: 
        print("Không có số nào")
        return
    while f1 + f2 <= a:
        f_next = f1 + f2
        f1 = f2
        f2 = f_next
    print(f"Số Fibonacci lớn nhất <= {a} là: {f2}")

# Bài 20: Đếm chữ số là số nguyên tố (2, 3, 5, 7)
def bai_20():
    n = int(input("Nhập n: ")) 
    count = 0
    temp = n
    while temp > 0:
        d = temp % 10
        if d in (2, 3, 5, 7):
            count += 1
        temp //= 10
    print(f"Số lượng chữ số là số nguyên tố: {count}")
```
