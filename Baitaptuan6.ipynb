```python
from typing import Tuple, List

# W5A1 & W5A13: Trả về số lớn hơn trong hai số
def Max_Of_Two(a: int, b: int) -> int:
    return a if a > b else b

# W5A2: Hoán đổi hai số nguyên
def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a

# W5A3: Kiểm tra số nguyên tố
def is_prime(n: int) -> bool:
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# W5A4: Kiểm tra số hoàn hảo
def is_perfect_number(n: int) -> bool:
    if n <= 0: return False
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n

# W5A5: Tìm vị trí đầu tiên của k trong list
def find_first_position(lst: List[int], k: int) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

# W5A6: Tính giai thừa n!
def factorial(n: int) -> int:
    if n < 0: return 0
    kq = 1
    for i in range(1, n + 1):
        kq *= i
    return kq
```

```python
# W5A7 & W5A14: Giả lập máy tính bỏ túi 
def compute(num1: float, num2: float, op: str) -> float:
    if op == '+': return round(num1 + num2, 2) 
    elif op == '-': return round(num1 - num2, 2) 
    elif op == '*': return round(num1 * num2, 2) 
    elif op == '/':
        if num2 == 0: return "Error: Division by zero"
        return round(num1 / num2, 2)
    return 0.0

# W5A11: Kiểm tra tích a*b có chia hết cho 10 không
def is_product_divisible_by_10(a: int, b: int) -> bool:
    return (a * b) % 10 == 0 

# W5A12: Chia lấy thương và dư 
def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    q = dividend // divisor 
    r = dividend % divisor 
    return q, r 

# W5A15: Kiểm tra tam giác vuông 
def is_right_triangle(a: float, b: float, c: float) -> bool:
    sides = sorted([a, b, c]) 
    return round(sides[0]**2 + sides[1]**2, 5) == round(sides[2]**2, 5)
```

```python
# W5A8: Khoảng cách Hamming giữa hai số nguyên 
def hamming_distance(x: int, y: int) -> int:
    xor_result = x ^ y
    distance = 0
    while xor_result > 0:
        distance += xor_result & 1
        xor_result >>= 1 
    return distance 

# W5A9: Tổng các chữ số của n 
def sum_of_digits(n: int) -> int:
    n = abs(n)
    tong = 0
    while n > 0:
        tong += n % 10 
        n //= 10 
    return tong 

# W5A10: Kiểm tra hai từ đẳng cấu (Isomorphic Strings) 
def are_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b): return False 
    
    mapping_a_to_b = {}
    mapping_b_to_a = {}
    
    for char_a, char_b in zip(a, b):
        # Kiểm tra ánh xạ từ a sang b
        if char_a in mapping_a_to_b:
            if mapping_a_to_b[char_a] != char_b: return False 
        else:
            mapping_a_to_b[char_a] = char_b 
            
        # Kiểm tra ánh xạ ngược từ b sang a để đảm bảo 1-1
        if char_b in mapping_b_to_a:
            if mapping_b_to_a[char_b] != char_a: return False 
        else:
            mapping_b_to_a[char_b] = char_a 
            
    return True 
```

```python
import math

# W5A16: Kiểm tra số Armstrong (ví dụ: 153 = 1^3 + 5^3 + 3^3)
def w5a16_is_armstrong(x: int) -> bool:
    if x < 0: return False
    s_x = str(x)
    n = len(s_x)
    tong = sum(int(digit)**n for digit in s_x)
    return tong == x

# W5A17: Kiểm tra số chính phương
def w5a17_is_perfect_square(x: int) -> bool:
    if x < 0: return False
    root = int(math.sqrt(x))
    return root * root == x

# W5A18: Kiểm tra số đối xứng
def w5a18_is_palindrome(n: int) -> bool:
    s = str(abs(n))
    return s == s[::-1]

# W5A19: Kiểm tra số kỳ diệu (Magic Number)
def w5a19_is_magic_number(n: int) -> bool:
    if n <= 0: return False
    while n >= 10:
        n = sum(int(d) for d in str(n))
    return n == 1

# W5A20: Nhận diện kiểu số
def detect_number_type(s: str) -> str:
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'

# W5A21: Một số bằng tích hai số còn lại
def is_one_product_of_others(a: float, b: float, c: float) -> bool:
    return (a*b == c) or (a*c == b) or (b*c == a)
```

```python
# W5A22: Tổng hai nửa của chuỗi 6 chữ số
def sum_halves(s: str) -> int:
    if len(s) == 6:
        return int(s[:3]) + int(s[3:]) 
    return 0

# W5A23: Đếm số nghiệm thực phương trình bậc 2
def quadratic_roots_count(a: float, b: float, c: float) -> int:
    if a == 0: 
        return 1 if b != 0 else (0 if c != 0 else -1) # -1 đại diện vô số nghiệm
    delta = b**2 - 4*a*c 
    if delta > 0: return 2 
    if delta == 0: return 1 
    return 0 

# W5A24: Kiểm tra 3 điểm thẳng hàng
def are_collinear(x1, y1, x2, y2, x3, y3) -> bool:
    # Sử dụng diện tích tam giác bằng 0
    return (x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) == 0 

# W5A25: Kiểm tra điểm p nằm trong tứ giác quad (4 đỉnh)
def point_in_quadrilateral(quad, p):
    # Sử dụng thuật toán Ray Casting đơn giản
    x, y = p 
    n = len(quad) 
    inside = False 
    p1x, p1y = quad[0] 
    for i in range(n + 1): 
        p2x, p2y = quad[i % n] 
        if y > min(p1y, p2y): 
            if y <= max(p1y, p2y): 
                if x <= max(p1x, p2x): 
                    if p1y != p2y: 
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x 
                    if p1x == p2x or x <= xinters:
                        inside = not inside 
        p1x, p1y = p2x, p2y 
    return inside 

# W5A26: Hàm XOR
def w5a26_xor(a, b):
    return a ^ b 
```

```python
# W5A27: Đếm chữ số khác 0 trong list
def count_nonzero(lst: list) -> int:
    return sum(1 for x in lst if x != 0)

# W5A28: Tính góc ABC (độ) từ 3 điểm
def calculate_angle(x1, y1, x2, y2, x3, y3):
    # Vector BA và BC
    ba = (x1-x2, y1-y2) 
    bc = (x3-x2, y3-y2) 
    dot_prod = ba[0]*bc[0] + ba[1]*bc[1] 
    mag_ba = math.sqrt(ba[0]**2 + ba[1]**2) 
    mag_bc = math.sqrt(bc[0]**2 + bc[1]**2) 
    cos_angle = dot_prod / (mag_ba * mag_bc) 
    return math.degrees(math.acos(max(-1, min(1, cos_angle)))) 

# W5A29: Max và Min của 3 số
def max_min_of_three(a, b, c):
    return [max(a, b, c), min(a, b, c)]

# W5A30: Hệ số nhị thức C(n, k)
def binomial_coefficient(n, k):
    if k < 0 or k > n: return 0 
    return math.comb(n, k) 

# W5A31: Tổng bình phương từ 1 đến n
def sum_squares(n: int) -> int:
    return sum(i**2 for i in range(1, n + 1)) 

# W5A32: Tính tổng bằng hàm eval
def eval_sum():
    expr = input("Nhập biểu thức (vd: 5+10): ") 
    return eval(expr) 

# W5A33: Tính tiền taxi thông minh
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Quy tắc giả định (vì đề không nêu chi tiết giá)
    base_price = 15000 if car_type == '4-seat' else 20000 
    price_per_km = 12000
    wait_fee = wait_minutes * 500 
    total = base_price + (km * price_per_km) + wait_fee 
    if start_hour >= 22 or start_hour <= 5: total *= 1.1 
    return round(total) 

# W5A34: Tổng các số x từ 1 đến n có các chữ số khác nhau
def sum_1_to_n_distinct(n: int) -> int:
    if n <= 0: return 0 
    total = 0
    for x in range(1, n + 1):
        s_x = str(x) 
        if len(set(s_x)) == len(s_x): 
            total += x 
    return total 
```

```python
import math

def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    """Bài 1: Tính cước phí Taxi thông minh [cite: 110]"""
    if car_type == 4:
        fare = 11000
        if km > 0.8:
            if km <= 20:
                fare += (km - 0.8) * 13000
            else:
                fare += (20 - 0.8) * 13000 + (km - 20) * 10500
    else:
        fare = 12000
        if km > 0.8:
            if km <= 20:
                fare += (km - 0.8) * 14000
            else:
                fare += (20 - 0.8) * 14000 + (km - 20) * 12000

    if wait_minutes > 5:
        wait_blocks = math.ceil((wait_minutes - 5) / 5)
        fare += wait_blocks * 2000
    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        fare *= 1.2

    if km > 30 and not is_night:
        fare *= 0.9
    if fare < 50000 and km > 5:
        fare = 50000
    return int(fare)
```

```python
# Bài 16: Tính góc ABC 
def angle_abc(x1, y1, x2, y2, x3, y3):
    # Vector BA và BC 
    ba = (x1-x2, y1-y2)
    bc = (x3-x2, y3-y2)
    dot = ba[0]*bc[0] + ba[1]*bc[1]
    m_ba = math.sqrt(ba[0]**2 + ba[1]**2)
    m_bc = math.sqrt(bc[0]**2 + bc[1]**2)
    cos_val = dot / (m_ba * m_bc)
    return math.degrees(math.acos(max(-1, min(1, cos_val))))

# Bài 17: Khoảng cách Euclid 
def euclid_distance(x1, y1, x2, y2):
    return math.sqrt((x1-x2)**2 + (y1-y2)**2)

# Bài 18: Đếm chữ số khác 0
def count_nonzero_digits(n):
    return sum(1 for d in str(abs(n)) if d != '0')

# Bài 19: Hàm XOR 
def bitwise_xor(a, b):
    return a ^ b

# Bài 20: Lũy thừa dư (a^b^c^d % mod) 
def complex_mod(a, b, c, d, mod):
    return pow(a, b * c * d, mod)

# Bài 22: Diện tích hình tròn 
def circle_area(r):
    pi = 3.14
    return pi * r * r

# Bài 25: Điểm trong tứ giác 
def point_in_quadrilateral(quad, p):
    n = len(quad)
    inside = False
    x, y = p
    p1x, p1y = quad[0]
    for i in range(n + 1):
        p2x, p2y = quad[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside

# Bài 28: Tổng hai nửa chuỗi 6 chữ số 
def sum_halves(s):
    if len(s) != 6 or not s.isdigit():
        raise ValueError("Sai độ dài hoặc chứa ký tự không phải số") 
    return int(s[:3]) + int(s[3:]) 

# Bài 29: Một số là tích hai số còn lại 
def is_one_product_of_others(a, b, c):
    return a*b == c or a*c == b or b*c == a 

# Bài 30: Nhận diện kiểu số 
def detect_number_type(s):
    if not s: return 'neither'
    if s[0] in '+-': s_check = s[1:]
    else: s_check = s
    if s_check.isdigit() and '.' not in s:
        return 'int' 
    if '.' in s:
        parts = s.split('.')
        if len(parts) == 2 and parts[0] != '' and parts[1] != '':
            if (parts[0].isdigit() or (parts[0][0] in '+-' and parts[0][1:].isdigit())) and parts[1].isdigit():
                return 'float'
    return 'neither' 
```
