```python
# W4A1: Tính tổng từ 1 đến n (n <= 1000)
def w4a1_sum_of_numbers():
    n = int(input("Nhập n: "))
    tong = 0
    for i in range(1, n + 1):
        tong += i
    print(tong) 

# W4A2: Kiểm tra số nguyên tố (Nhập cho đến khi dương mới xử lý)
def w4a2_prime_number():
    while True:
        n = int(input("Nhập một số nguyên dương: "))
        if n > 0: break
    
    if n < 2:
        print("Không phải số nguyên tố")
    else:
        is_prime = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
        print("Là số nguyên tố" if is_prime else "Không phải số nguyên tố") 

# W4A3: Tính giai thừa n! (0 < n < 100)
def w4a3_factorial():
    n = int(input("Nhập n: "))
    kq = 1
    for i in range(1, n + 1):
        kq *= i
    print(kq)

# W4A4: Đếm số chữ số (trừ dấu, không dùng string)
def w4a4_counting_digits():
    n = abs(int(input("Nhập n: ")))
    if n == 0:
        count = 1
    else:
        count = 0
        while n > 0:
            count += 1
            n //= 10
    print(count)
```

```python
# W4A5 & W4A6: Tìm ý nghĩa cuộc đời (số 42)
def w4a5_6_meaning_of_life():
    n = int(input("Nhập số lượng phần tử n: "))
    found = False
    for i in range(n):
        x = int(input(f"Nhập số thứ {i+1}: "))
        if x == 42:
            found = True
    print("I've found the meaning of life!" if found else "It's a joke!") 

# W4A7 - Phần 1: Tổng các số nguyên tố trong đoạn [a, b]
def w4a7_sum_primes_in_range():
    a, b = map(int, input().split())
    tong = 0
    for num in range(a, b + 1):
        if num < 2: continue
        is_p = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_p = False
                break
        if is_p: tong += num
    print(tong) 

# W4A7 - Phần 2: Ước số nguyên tố lớn nhất của n
def w4a7_largest_prime_factor():
    n = int(input("Nhập n (>=2): "))
    max_prime = -1
    d = 2
    while d * d <= n:
        while n % d == 0:
            max_prime = d
            n //= d
        d += 1
    if n > 1: max_prime = n
    print(max_prime) 
```

```python
# W4A8: Reverse-Add Palindrome
def w4a8_palindrome():
    n = int(input("Nhập n: "))
    steps = 0
    while True:
        s_n = str(n)
        if s_n == s_n[::-1]:
            print(f"{steps} {n}")
            break
        n += int(s_n[::-1])
        steps += 1 

# W4A9: Số chính phương có các chữ số đôi một khác nhau
def w4a9_distinct_digits_square():
    n_limit = int(input("Nhập n: "))
    results = []
    for i in range(1, int(n_limit**0.5) + 1):
        sq = i * i
        if sq > n_limit: break
        s_sq = str(sq)
        if len(set(s_sq)) == len(s_sq): 
            results.append(str(sq))
    
    if results:
        print(" ".join(results))
    else:
        print("(no number)") 

# W4A10: Dãy Collatz dài nhất trong khoảng [1, n]
def w4a10_longest_collatz():
    n = int(input("Nhập n: "))
    max_len = 0
    best_x = 1
    for i in range(1, n + 1):
        curr = i
        length = 1
        while curr != 1:
            if curr % 2 == 0: curr //= 2
            else: curr = 3 * curr + 1
            length += 1
        if length > max_len:
            max_len = length
            best_x = i
    print(f"{best_x} {max_len}") 

# W4A11: Đếm số lượng ước số chẵn của n
def w4a11_even_divisors():
    n = int(input("Nhập n: "))
    count = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i % 2 == 0: count += 1
            if (n // i) != i and (n // i) % 2 == 0:
                count += 1
    print(count) 
```

```python
# W4A12: Tính lãi suất ngân hàng (0.7%/tháng)
def w4a12_interest():
    x = float(input("Nhập số tiền gửi X: "))
    n = int(input("Nhập số tháng N: "))
    for _ in range(n):
        x += x * 0.007
    print(int(x)) 

# W4A13: Cặp số thân thiết (Amicable numbers)
def w4a13_close_pair():
    def sum_divisors(num):
        s = 0
        for i in range(1, num // 2 + 1):
            if num % i == 0: s += i
        return s
    
    a = int(input("Nhập số thứ nhất: "))
    b = int(input("Nhập số thứ hai: "))
    if sum_divisors(a) == b and sum_divisors(b) == a:
        print("true")
    else:
        print("false") 

# W4A14: Tìm ước chung lớn nhất (GCD)
def w4a14_gcd():
    m, n = map(int, input().split())
    while n:
        m, n = n, m % n
    print(m) 

# W4A15: Bài toán Chó và Gà (Mở rộng)
def w4a15_chicken_and_dog():
    tong_con = int(input("Nhập tổng số con: "))
    tong_chan = int(input("Nhập tổng số chân: "))
    found = False
    for x in range(tong_con + 1):
        y = tong_con - x
        if 2*x + 4*y == tong_chan:
            print(f"Gà: {x}, Chó: {y}")
            found = True
            break
    if not found:
        print("invalid")
```

```python
# W4A16: In các số chẵn chia hết cho 3 bé hơn 100 
def w4a16():
    for i in range(1, 100):
        if i % 2 == 0 and i % 3 == 0:
            print(i, end=" ")
    print()

# W4A17: In bảng cửu chương của a 
def w4a17():
    a = int(input("Nhập a: "))
    for i in range(1, 11):
        print(f"{a} x {i} = {a * i}")

# W4A18: In toàn bộ ước chung của a và b 
def w4a18():
    a = int(input("Nhập a: "))
    b = int(input("Nhập b: "))
    limit = a if a < b else b
    for i in range(1, limit + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
    print()

# W4A19: In số dương chẵn <= n (n <= 1000) 
def w4a19():
    n = int(input("Nhập n: "))
    for i in range(2, n + 1, 2):
        print(i, end=" ")
    print()

# W4A20: Kiểm tra n có dạng 2^k hay không 
def w4a20():
    n = int(input("Nhập n: "))
    if n <= 0:
        print("Không phải dạng 2^k")
        return
    temp = n
    while temp % 2 == 0:
        temp //= 2
    if temp == 1:
        print(f"{n} là số dạng 2^k")
    else:
        print(f"{n} không phải dạng 2^k")
```

```python
# W4A21: Tính tổng các chữ số của n 
def w4a21():
    n = abs(int(input("Nhập n: ")))
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    print(f"Tổng các chữ số: {tong}")

# W4A22: Đếm chữ số chẵn và lẻ 
def w4a22():
    n = abs(int(input("Nhập n: ")))
    chan = le = 0
    if n == 0: chan = 1
    while n > 0:
        if (n % 10) % 2 == 0: chan += 1
        else: le += 1
        n //= 10
    print(f"Chẵn: {chan}, Lẻ: {le}")

# W4A23: Tìm k sao cho S(k) = 1+2+...+k < n lớn nhất
def w4a23():
    n = int(input("Nhập n: "))
    s = 0
    k = 0
    while s + (k + 1) < n:
        k += 1
        s += k
    print(f"k lớn nhất là: {k} (S = {s})")

# W4A24: Tìm n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n > A
def w4a24():
    a = float(input("Nhập A: "))
    s = 0
    n = 0
    while s <= a:
        n += 1
        s += 1/n
    print(f"n nhỏ nhất là: {n}")

# W4A25: Tìm Max/Min trong dãy dừng bởi -1 (Không dùng mảng) 
def w4a25():
    print("Nhập dãy số (nhập -1 để dừng):")
    n = int(input())
    if n == -1: return
    max_val = min_val = n
    while True:
        n = int(input())
        if n == -1: break
        if n > max_val: max_val = n
        if n < min_val: min_val = n
    print(f"Max: {max_val}, Min: {min_val}")

# W4A26: Số Fibonacci lớn nhất không vượt quá A 
def w4a26():
    a = int(input("Nhập A: "))
    f1, f2 = 1, 1
    if a < 1: return
    while f1 + f2 <= a:
        f1, f2 = f2, f1 + f2
    print(f"Số Fibonacci lớn nhất <= {a} là: {f2}")
```

```python
# W4A27: Đếm số từ (Không dùng split, len)
def w4a27():
    s = input("Nhập chuỗi: ")
    count = 0
    in_word = False
    # Duyệt qua từng ký tự bằng index
    for i in range(2000): 
        try:
            char = s[i]
            if char != " " and not in_word:
                count += 1
                in_word = True
            elif char == " ":
                in_word = False
        except IndexError:
            break
    print(f"Số từ: {count}")

# W4A28: In từ đầu tiên (Không dùng split) 
def w4a28():
    s = input("Nhập chuỗi: ")
    first_word = ""
    start = False
    for i in range(2000):
        try:
            char = s[i]
            if char != " ":
                first_word += char
                start = True
            elif start: 
                break
        except IndexError:
            break
    print(f"Từ đầu tiên: {first_word}")

# W4A29: Đổi hoa - thường 
def w4a29():
    s = input("Nhập chuỗi: ")
    res = ""
    for char in s:
        if 'a' <= char <= 'z':
            res += chr(ord(char) - 32)
        elif 'A' <= char <= 'Z':
            res += chr(ord(char) + 32)
        else:
            res += char
    print(res)

# W4A30: Tính tổng 3 số cách nhau bởi dấu phẩy 
def w4a30():
    s = input("Nhập 3 số (vd: 3, 12, 15): ")
    tong = 0
    current_num = ""
    for char in s:
        if char != "," and char != " ":
            current_num += char
        else:
            if current_num:
                tong += int(current_num)
                current_num = ""
    if current_num: 
        tong += int(current_num)
    print(f"Tổng: {tong}")

# W4A31: Đếm ký tự hoa, thường, số 
def w4a31():
    s = input("Nhập chuỗi: ")
    hoa = thuong = so = 0
    for char in s:
        if 'A' <= char <= 'Z': hoa += 1
        elif 'a' <= char <= 'z': thuong += 1
        elif '0' <= char <= '9': so += 1
    print(f"Hoa: {hoa}, Thường: {thuong}, Số: {so}")
```
