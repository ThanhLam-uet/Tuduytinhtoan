```python
import math

# Bài 1: Diện tích và chu vi hình chữ nhật 
def bai_1():
    a = float(input("Nhập cạnh a: "))
    b = float(input("Nhập cạnh b: "))
    print(f"Chu vi: {(a + b) * 2}")
    print(f"Diện tích: {a * b}")
```

```python
# Bài 3: Kiểm tra và tính toán tam giác 
def bai_3():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    c = float(input("Nhập c: "))
    
    if a + b > c and a + c > b and b + c > a:
        if a == b == c:
            loai = "Tam giác đều"
        elif a == b or b == c or a == c:
            if (a**2 + b**2 == round(c**2, 5)) or (a**2 + c**2 == round(b**2, 5)) or (b**2 + c**2 == round(a**2, 5)):
                loai = "Tam giác vuông cân"
            else:
                loai = "Tam giác cân"
        elif (a**2 + b**2 == round(c**2, 5)) or (a**2 + c**2 == round(b**2, 5)) or (b**2 + c**2 == round(a**2, 5)):
            loai = "Tam giác vuông"
        else:
            loai = "Tam giác thường"
        chu_vi = a + b + c
        p = chu_vi / 2
        dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
        
        print(f"Đây là: {loai}")
        print(f"Chu vi: {chu_vi}, Diện tích: {dien_tich:.2f}")
    else:
        print("Khong la ba canh cua mot tam giac")
```

```python
# Bài 4: Giải phương trình bậc 1 ax + b = 0 
def bai_4():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    if a == 0:
        print("Vô nghiệm" if b != 0 else "Vô số nghiệm")
    else:
        print(f"Nghiệm x = {-b/a}")

# Bài 5: Giải phương trình bậc 2 ax^2 + bx + c = 0 
def bai_5():
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    c = float(input("Nhập c: "))
    if a == 0:
        # Trở thành pt bậc 1
        if b == 0:
            print("Vô nghiệm" if c != 0 else "Vô số nghiệm")
        else:
            print(f"Nghiệm x = {-c/b}")
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
            print("Phương trình vô nghiệm")
        elif delta == 0:
            print(f"Nghiệm kép x = {-b/(2*a)}")
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print(f"x1 = {x1}, x2 = {x2}")

# Bài 8: Giải hệ phương trình ax + by = m và cx + dy = n 
def bai_8():
    a = float(input("a: ")); b = float(input("b: ")); m = float(input("m: "))
    c = float(input("c: ")); d = float(input("d: ")); n = float(input("n: "))
    D = a*d - b*c
    Dx = m*d - b*n
    Dy = a*n - m*c
    if D != 0:
        print(f"Hệ có nghiệm duy nhất: x = {Dx/D}, y = {Dy/D}")
    else:
        if Dx == 0 and Dy == 0:
            print("Hệ vô số nghiệm")
        else:
            print("Hệ vô nghiệm")
```

```python
# Bài 6 & 7: Max/Min của 4 số 
def bai_6_7():
    nums = [float(input(f"Nhập số thứ {i+1}: ")) for i in range(4)]
    print(f"Max: {max(nums)}")
    print(f"Min: {min(nums)}")

# Bài 9: Chuyển đổi giây sang Giờ:Phút:Giây 
def bai_9():
    x = int(input("Nhập số giây: "))
    gio = x // 3600
    phut = (x % 3600) // 60
    giay = x % 60
    print(f"{gio} giờ, {phut} phút, {giay} giây")

# Bài 10: Kiểm tra điểm A(x,y) có thuộc đường tròN
def bai_10():
    x0 = float(input("Tâm x0: "))
    y0 = float(input("Tâm y0: "))
    r = float(input("Bán kính r: "))
    xa = float(input("Điểm A - x: "))
    ya = float(input("Điểm A - y: "))
    
    khoang_cach = math.sqrt((xa - x0)**2 + (ya - y0)**2)
    if khoang_cach == r:
        print("Điểm A nằm trên đường tròn")
    elif khoang_cach < r:
        print("Điểm A nằm trong đường tròn")
    else:
        print("Điểm A nằm ngoài đường tròn")

# Bài 11: Tính x^y
def bai_11():
    x = float(input("Nhập x: "))
    y = float(input("Nhập y: "))
    print(f"Cách 1: {x ** y}") 
```

```python
# W3A1: Đảo ngược số nguyên 3 chữ số (không dùng chuỗi)
def w3a1_reverse():
    n = int(input("Nhập số nguyên dương có 3 chữ số: "))
    if 100 <= n <= 999:
        hang_tram = n // 100 
        hang_chuc = (n // 10) % 10 
        hang_don_vi = n % 10
        n_dao_nguoc = hang_don_vi * 100 + hang_chuc * 10 + hang_tram 
        print(n_dao_nguoc)
    else:
        pass 

# W3A2: Hoán đổi 2 số dùng XOR (không dùng biến tạm)
def w3a2_xor_swap():
    a = int(input("Nhập a: "))
    b = int(input("Nhập b: "))
    a = a ^ b 
    b = a ^ b 
    a = a ^ b 
    print(f"Sau khi hoán đổi: a = {a}, b = {b}")

# W3A3: Kiểm tra lũy thừa của 2 bằng Bitwise
def w3a3_power_of_2():
    n = int(input("Nhập n: "))
    if n > 0 and (n & (n - 1) == 0): 
        print("Phải là lũy thừa của 2")
    else:
        print("Không phải lũy thừa của 2")

# W3A4: Chia m cho n làm tròn xuống
def w3a4_floor():
    m, n = map(int, input().split())
    print(m // n) 

# W3A5: Chia m cho n làm tròn lên
def w3a5_ceil():
    m, n = map(int, input().split())
    # Công thức làm tròn lên không dùng thư viện: (m + n - 1) // n
    print((m + n - 1) // n) 
```

```python
# W3A6: Kiểm tra chẵn lẻ
def w3a6_even_odd():
    x = int(input())
    print("Even" if x % 2 == 0 else "Odd")

# W3A7: Kiểm tra hai số cùng âm
def w3a7_check_negative():
    a, b = map(int, input().split())
    if a < 0 and b < 0: 
        print("Yes")
    else:
        print("No")

# W3A8: So sánh độ dài chuỗi
def w3a8_longer_string():
    a = input()
    b = input()
    print("True" if len(a) > len(b) else "False")

# W3A9: Kiểm tra tính hợp lệ của tam giác
def w3a9_triangle():
    a, b, c = map(int, input().split())
    if a + b > c and a + c > b and b + c > a: 
        print("Yes")
    else:
        print("No")

# W3A10: Tìm số lớn nhất trong 4 số
def w3a10_largest():
    nums = list(map(int, input().split()))
    max_val = nums[0]
    for n in nums:
        if n > max_val:
            max_val = n
    print(max_val)
```

```python
# W3A11: Phân loại tam giác
def w3a11_classify_triangle():
    a, b, c = map(int, input().split())
    if a + b > c and a + c > b and b + c > a: 
        if a == b == c:
            print("Tam giác đều") 
        elif a == b or b == c or a == c:
            print("Tam giác cân") 
        else:
            print("Tam giác thường") 
    else:
        print("Không phải tam giác") 

# W3A12: Kiểm tra năm nhuận
def w3a12_leap_year():
    n = int(input())
    if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0): 
        print("Yes")
    else:
        print("No")

# W3A14: Giải phương trình bậc nhất ax + b = 0
def w3a14_linear_equation():
    a, b = map(float, input().split())
    if a == 0:
        print("Vô số nghiệm" if b == 0 else "Vô nghiệm") 
    else:
        print(f"{-b/a:.2f}") 

# W3A15: Xếp loại học lực
def w3a15_grading():
    dtb = float(input())
    if dtb >= 8.0:
        print("Gioi") 
    elif dtb >= 6.5:
        print("Khá") 
    elif dtb >= 5.0:
        print("Trung bình")
    else:
        print("Yếu") 

# W3A16: Làm tròn (không dùng hàm có sẵn)
def w3a16_manual_round():
    n = float(input())
    # Tròn xuống
    tron_xuong = int(n) if n >= 0 else int(n) - 1
    # Tròn lên
    tron_len = int(n) + 1 if n > int(n) else int(n)
    # Tròn gần nhất
    gan_nhat = int(n + 0.5) if n >= 0 else int(n - 0.5)
    print(f"{tron_len} {tron_xuong} {gan_nhat}") 

# W3A17: Kiểm tra cấp số nhân
def w3a17_geometric_progression():
    a, b, c, d = map(int, input().split())
    if a != 0 and b % a == 0:
        q = b // a 
        if b * q == c and c * q == d: 
            print("Yes")
            return
    print("No")
```
